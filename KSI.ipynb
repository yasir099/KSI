We can observe that there is a significant percentage of missing data in a particular column, represented by ACCNUM (Accident Number). Each unique row in the ACCNUM column represents the number of accidents. Approximately 27% of the ACCNUM data is missing, and to address this, we can generate a unique number for each accident by generating a number for each match in the YEAR, LATITUDE, and TIME rows. This approach allows us to fill in the missing values and ensure that each accident has a distinct identifier.

```python
# Identify the rows with missing values in the "ACCNUM" column
missing_accnum = data['ACCNUM'].isnull()

# Group the rows by their common values in the "LONGITUDE," "LATITUDE," and "YEAR" columns
grouped = data.groupby(['LONGITUDE', 'LATITUDE', 'YEAR'])

# Assign unique numbers to each group of rows with the same common values
unique_numbers = grouped.ngroup() + 1

# Replace the missing values in the "ACCNUM" column with the generated unique numbers
data.loc[missing_accnum, 'ACCNUM'] = unique_numbers[missing_accnum]
```

Drop some columns 

```python
data.drop([ 'OFFSET','PEDTYPE', 'PEDACT', 'PEDCOND', 'CYCLISTYPE', 'CYCACT', 'CYCCOND', 'X', 'Y'], axis=1, inplace=True)
```
